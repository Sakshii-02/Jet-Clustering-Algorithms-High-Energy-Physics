```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import awkward as ak
```

```python
file=uproot.open("OutFile_data.root")
tree=file["demo/Events"]
branches=tree.arrays()
branches[1].tolist()
```

```python
plt.hist(branches['nPFJets'], bins=10, range=(0, 30))
plt.xlabel('# of PFjets')
plt.ylabel('Events')
plt.show()
```

```python
plt.hist(ak.flatten(branches['PFJet_M']), bins=25, range=(0, 50))
plt.xlabel('$m_{\mathrm{Jet}}$ [GeV]')
plt.ylabel('Number of jets / 2 GeV')
plt.show()
```

```python
plt.hist([ak.flatten(branches['PFJet_M']),ak.flatten(branches['PuppiJet_M'])], label=["AK4PFJets","AK4PuppiJets"], bins=25, range=(0, 50))
plt.xlabel('$m_{\mathrm{Jet}}$ [GeV]')
plt.ylabel('Number of jets / 2 GeV')
plt.legend()
plt.show()
```

```python
plt.hist(branches['nPV'], bins=30, range=(0, 30))
plt.xlabel('# of PVs')
plt.ylabel('Events')
plt.show()
```

```python
plt.scatter(branches['nPV'],branches['nPFJets'])
plt.xlabel('# of PVs')
plt.ylabel('# of PFJets')
plt.show()
```

```python
plt.scatter(branches['nPV'],branches['puppiMET'])
plt.xlabel('# of PVs')
plt.ylabel('Puppi MET [GeV]')
plt.show()
```

```python
plt.hist(branches['npfCands'], bins=100, range=(0, 500))
plt.xlabel('# of PF cands')
plt.ylabel('Events')
plt.show()
```

```python
plt.hist([ak.flatten(branches["pfCand_PuppiWeight"]), ak.flatten(branches["pfCand_PuppiWeightNoLep"])], label=["PuppiWeight","PuppiWeightNoLep"], bins=10, range=(0, 1))
plt.xlabel('$w_{\mathrm{Puppi}}$')
plt.ylabel('# of PF cands')
plt.yscale('log')
plt.legend()
plt.show()
```
